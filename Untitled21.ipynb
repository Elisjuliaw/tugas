```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# üìå Upload gambar
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Mengambil nama file gambar

# üìå Membaca gambar dalam format grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# üìå Fungsi untuk menampilkan gambar
def show_images(images, titles):
    plt.figure(figsize=(15, 5))
    for i in range(len(images)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i], cmap="gray")
        plt.title(titles[i])
        plt.axis("off")
    plt.show()

# üìå Cek apakah gambar berhasil dimuat
if image is None:
    print("‚ùå Error: Gambar tidak ditemukan atau format tidak didukung!")
else:
    print("‚úÖ Gambar berhasil dimuat!")

    # 1. Citra Negatif
    def negative_image(img):
        return 255 - img

    # 2. Log Transformasi
    def log_transform(img):
        c = 255 / np.log(1 + np.max(img))
        return np.array(c * np.log(1 + img), dtype=np.uint8)

    # 3. Transformasi Power-Law (Gamma Correction)
    def gamma_correction(img, gamma=1.5):
        c = 255 / np.max(img)
        return np.array(c * (img / 255) ** gamma * 255, dtype=np.uint8)

    # 4. Pemerataan Histogram
    def histogram_equalization(img):
        return cv2.equalizeHist(img)

    # 5. Normalisasi Histogram
    def normalize_histogram(img):
        return cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

    # 6. Konversi RGB ke HSI + Thresholding
    def rgb_to_hsi_threshold(image_path, threshold=100):
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Konversi ke RGB

        # Konversi ke grayscale untuk mendapatkan intensitas (I)
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

        # Thresholding pada intensitas (I)
        _, thresholded = cv2.threshold(gray, threshold, 255, cv2.THRESH_BINARY)

        # Menampilkan hasil
        show_images([img, thresholded], ["RGB Image", "Thresholded Image"])

    # üìå Menerapkan transformasi
    negative = negative_image(image)
    log_transformed = log_transform(image)
    gamma_corrected = gamma_correction(image, gamma=1.5)
    hist_equalized = histogram_equalization(image)
    normalized = normalize_histogram(image)

    # üìå Menampilkan hasil transformasi
    show_images(
        [image, negative, log_transformed, gamma_corrected, hist_equalized, normalized],
        ["Original", "Negative", "Log Transform", "Gamma Correction", "Histogram Equalization", "Histogram Normalization"]
    )

    # üìå Uji konversi RGB ke HSI dan thresholding
    rgb_to_hsi_threshold(image_path, threshold=100)
```
