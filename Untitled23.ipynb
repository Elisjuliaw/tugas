```python
# Untuk upload file dari lokal di Google Colab
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Upload gambar dari komputer
uploaded = files.upload()

# Ambil nama file yang di-upload
filename = list(uploaded.keys())[0]

# Fungsi segmentasi Otsu 3 kelas
def segmentasi_otsu_3kelas(gambar_path):
    try:
        # Baca gambar dalam grayscale
        gambar = cv2.imread(gambar_path, cv2.IMREAD_GRAYSCALE)
        if gambar is None:
            raise ValueError("Gagal membaca gambar. Format tidak didukung atau file corrupt.")

        # Resize agar seragam
        gambar = cv2.resize(gambar, (256, 256))

        # Threshold Otsu pertama
        T1, _ = cv2.threshold(gambar, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        bawah = gambar[gambar < T1]
        atas = gambar[gambar >= T1]

        T2 = cv2.threshold(bawah, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[0] if len(bawah) > 0 else 0
        T3 = cv2.threshold(atas, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[0] if len(atas) > 0 else 255

        print(f"Thresholds Otsu: T1 = {T1:.2f}, T2 = {T2:.2f}, T3 = {T3:.2f}")

        # Segmentasi 3 kelas
        hasil = np.zeros_like(gambar)
        hasil[gambar < T2] = 85
        hasil[(gambar >= T2) & (gambar < T3)] = 170
        hasil[gambar >= T3] = 255

        # Tampilkan hasil
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 3, 1), plt.imshow(gambar, cmap='gray'), plt.title("Gambar Asli")
        plt.subplot(1, 3, 2), plt.hist(gambar.ravel(), 256), plt.title("Histogram")
        plt.subplot(1, 3, 3), plt.imshow(hasil, cmap='gray'), plt.title("Segmentasi Otsu 3 Kelas")
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"[âœ—] Terjadi kesalahan: {e}")

# Panggil fungsi
segmentasi_otsu_3kelas(filename)
```
